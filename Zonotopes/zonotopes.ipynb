```python
# import torch
import numpy as np
from zonotopes import Zonotope
```

```python
center = np.array([1., 1.])  # Center of the zonotope
generators = np.array([[1., 0.], [0., 1.],[1,1]])  # Generators of the zonotope

zonotope1 = Zonotope(center, generators)
zonotope2 = Zonotope(np.array([1., 1.]),np.array([[1., 0.], [1., 1.]]))
zonotope = zonotope1 + zonotope2
```

```python
zonotope.plot()
zonotope1.plot()
zonotope2.plot()
```
